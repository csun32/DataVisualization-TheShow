```python
fees = pd.read_csv('cancelfees.csv')

fees.info()
```

```python
print(fees.columns)
```

```python
fees
```

```python
# #define values
# values = ['Alaska', 'Sun Country', ' Sun Country' 'Spirit', 'Frontier', 'Allegiant', ' Alaska']

# #drop rows that contain any value in the list
# fees1 = fees[fees[' Airline '].isin(values) == False]
fees1 = fees.iloc[:-2]
fees1 = fees1.rename({' Airline ' : 'Airline'}, axis = 1)
fees1 = fees1.replace(',', '', regex = True)
fees1 = fees1.melt(id_vars = 'Airline', var_name = 'year', value_name = 'fees').sort_values('year')
fees1['fees'] = fees1['fees'].astype(int)
fees1['Airline'] = fees1['Airline'].str.strip()
fees1
```

```python
plt.figure(figsize=(11,7))
ax1 = sns.lineplot(data=fees1, x="year", y="fees", hue="Airline")

# Customize the axes and title
ax1.set_title("Reservation Cancellation/Change Fees by Airline")
ax1.set_xlabel("Year")
ax1.set_ylabel("Dollars in Millions")
ax1.set_yticklabels([x for x in (ax1.get_yticks() * 1000)/1000000])
```

```python
top_5 = ['United', 'Alaska', 'American', 'JetBlue', 'Delta']
fees_top5 = fees1[fees1['Airline'].isin(top_5)]

fees_top5['Airline'].unique()
```

```python
plt.figure(figsize=(11,7))
ax2 = sns.lineplot(data=fees_top5, x="year", y="fees", hue="Airline")

# Customize the axes and title
ax2.set_title("Reservation Cancellation/Change Fees by Top 5 Airlines with Highest Fees")
ax2.set_xlabel("Year")
ax2.set_ylabel("Dollars in Millions")
ax2.set_yticklabels([x for x in (ax1.get_yticks() * 1000)/1000000])
```

```python
rest = fees1[~fees1['Airline'].isin(top_5)]

rest['Airline'].unique()
```

```python
plt.figure(figsize=(11,7))
ax2 = sns.lineplot(data=rest, x="year", y="fees", hue="Airline")

# Customize the axes and title
ax2.set_title("Reservation Cancellation/Change Fees by Airlines without Top 5 Airlines with Highest Fees")
ax2.set_xlabel("Year")
ax2.set_ylabel("Dollars in Millions")
ax2.set_yticklabels([x for x in (ax1.get_yticks() * 1000)/1000000])
```
